```python
#% de données manquantes
def missing(df,detail=False):
    total = 0
    for col in df.columns:
        miss = df[col].isnull().sum()
        pct = df[col].isna().mean() * 100
        total += miss
        print('{} => {} [{}%]'.format(col, miss, round(pct, 2)))
        total=0
```
